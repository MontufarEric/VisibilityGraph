```python
def genConway(N):
    conway=[0,1,1]
    test=[]
    for n in range(3,N):
        an=conway[conway[n-1]]+conway[n-conway[n-1]]
        conway.append(an)
        test.append(an/n)
    return(test)
```

```python
x=genConway(1000)
plt.plot(x)
```
